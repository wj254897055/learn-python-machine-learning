```python
## 更高的方差意味着分布更广
high_var_array = np.array([1,100,200,300,4000,5000])
low_var_array = np.array([2,4,6,8,10])
print(f"high_var_array的方差为：{np.var(high_var_array)}")
print(f"low_var_array的方差为：{np.var(low_var_array)}")
print(f"high_var_array的标准差为：{np.std(high_var_array)}")
print(f"low_var_array的标准差为：{np.std(low_var_array)}")
print(f"high_var_array的平均值为：{np.mean(high_var_array)}")
print(f"low_var_array的平均值为：{np.mean(low_var_array)}")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(high_var_array)
plt.show()
plt.hist(low_var_array)
plt.show()
```

```python
price_table = pd.DataFrame(price_array,index=["price"],columns=["Apples","Oranges","Bananas"])
price_table
```
