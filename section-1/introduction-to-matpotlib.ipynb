```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro-') ##绘制一个简单的折线图，x轴为[1,2,3,4]，y轴为[1,4,9,16]，点的颜色为红色，线条为实线
```

```python
fig=plt.figure() ##创建一个新的图形对象
ax=fig.add_subplot() ##在图形对象中添加一个子图
ax.plot([1,2,3,4], [1,4,9,16], 'ro-') ##在子图中绘制一个简单的折线图，x轴为[1,2,3,4]，y轴为[1,4,9,16]，点的颜色为红色，线条为实线
```

```python
fig = plt.figure()  ##创建一个新的图形对象
ax = fig.add_axes([1,2,3,4])  ##在图形对象中添加一个子图
ax.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro-')
plt.show()##在子图中绘制一个简单的折线图，x轴为[1,2,3,4]，y轴为[1,4,9,16]，点的颜色为红色，线条为实线
```

```python
x = np.linspace(0,10,100) ##创建一个从0到10的等差数列，包含100个元素
y = np.sin(x) ##计算x的正弦值
fig, ax = plt.subplots() ##创建一个新的图形对象和一个子图对象
ax.plot(x,y) ##在子图中绘制x和y的关系图
```
