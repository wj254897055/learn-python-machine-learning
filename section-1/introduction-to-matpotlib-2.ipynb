```python
### 导入图表模块
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"] = ["Microsoft YaHei", "SimHei", "Arial Unicode MS"]
plt.rcParams["axes.unicode_minus"] = False
### 准备数据
x  = [1,2,3,4]
y = [1,4,9,16]
### 设置图表
fig, ax = plt.subplots(figsize=(10,10))
### 载入数据
ax.plot(x,y)
### 自定义设置
ax.set_title("图表样例")
ax.set_xlabel("x轴")
ax.set_ylabel("y轴")
fig.tight_layout()
fig.savefig("../images/样例图表.png")

```

## 通过Numpy生成数据 创建线图
* 折线图
* 散点图
* 柱状图
* 饼图
* 箱线图

```python
import  numpy as np
x = np.linspace(0,10,100)
## 生成正弦函数数据
fig, ax = plt.subplots(figsize=(10,10))
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x,x**2)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(x,np.exp(x))
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(x,np.sin(x))
```

```python
## 从字典创建柱状图
people_area = {"北京": 100, "上海": 80, "广州": 60, "深圳": 40}
fig, ax = plt.subplots(figsize=(10,10))
ax.bar(people_area.keys(), people_area.values())
ax.set_xlabel("城市")
ax.set_ylabel("人口密度")
ax.set_title("中国主要城市人口密度")
```
