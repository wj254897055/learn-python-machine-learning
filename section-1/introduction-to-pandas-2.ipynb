```python
car_sales_missing = pd.read_csv("../data/car-sales-missing-data.csv")
```

```python
car_sales_missing ##空值就会显示为NAN
```

```python
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean()) ##用Odometer (KM)列的平均值填充Odometer列中的空值
```

```python
car_sales_missing
```

```python
car_sales_missing.fillna(
    {"Odometer": car_sales_missing["Odometer"].mean()},
    inplace=True
) ##用Odometer (KM)列的平均值填充Odometer列
```

```python
car_sales_missing
```

### 忽略缺失值

```python
car_sales_missing.dropna(inplace=True) ##删除包含空值的行
```

```python
car_sales_missing
```

```python
car_sales_missing_drop = car_sales_missing.dropna()
```

```python
car_sales_missing_drop.to_csv("../data/car-sales-missing-data-dropped.csv") ##将删除空值后的数据保存到新的CSV文件中
```

### 创建新的数据

```python
seat_column = pd.Series([5,5,5,5,5]) ##创建一个新的Series对象，表示每辆车的座位数
```

```python
car_sales["Seats"] = seat_column
```

```python
car_sales.fillna({
    "Seats":5
}, inplace=True)
```

```python
car_sales ##将Seats列中的空值填充为5
```

### 创建一个新的列，表示每辆车的燃油经济性

```python
fuel_economy = [3.3,7.2,5.5,6.0,4.8,5.5,4.3,8.3,9,4] ##创建一个新的列表，表示每辆车的燃油经济性
car_sales["Fuel Economy"] = fuel_economy
```

```python
car_sales
```

```python
car_sales["Fuel Used"] = car_sales["Odometer (KM)"] /100 *  car_sales["Fuel Economy"] ##创建一个新的列，表示每辆车的燃油使用量
```

```python
car_sales
```
